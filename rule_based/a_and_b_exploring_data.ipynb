JAPANESE


```python
from transformers import BertTokenizer
from collections import Counter
import re

# Load a Japanese BERT tokenizer
tokenizer = BertTokenizer.from_pretrained("cl-tohoku/bert-base-japanese")

# Sample Japanese text (assuming you have a DataFrame with a 'question' column)
questions_ja = japanese_df['question'].tolist()

# Join all questions into a single string
text = ' '.join(questions_ja)

# Tokenize the text
tokens = tokenizer.tokenize(text)

# Define a function to filter tokens
def is_valid_word(token):
    # Filter out subwords, special tokens, and non-Japanese characters
    return not (token.startswith('##') or token in ['[CLS]', '[SEP]', '[PAD]', '[UNK]']) and re.match(r'^[ぁ-んァ-ン一-龯ａ-ｚＡ-Ｚ０-９々〆〤ー]+$', token) is not None

# Apply the filter to remove invalid tokens
filtered_words = [token for token in tokens if is_valid_word(token)]

# Count the frequency of the filtered words
word_freq = Counter(filtered_words)

# Get the most common words
most_common_words = word_freq.most_common(10)

# Print the most common words and their frequencies
print("Most common words:", most_common_words)

# Example translation (using Googletrans or an alternative)
from googletrans import Translator

translator = Translator()

# Translate the most common words
translations = [(word, translator.translate(word, src='ja', dest='en').text) for word, _ in most_common_words]

# Print the translations
for word, translation in translations:
    print(f"{word} -> {translation}")
```

```python
# Install necessary libraries
# !pip install mecab-python3 unidic-lite googletrans==4.0.0-rc1

import MeCab
from collections import Counter
from googletrans import Translator

# Initialize MeCab with word segmentation mode (Owakati)
mecab = MeCab.Tagger("-Owakati")  # Owakati mode provides whole words without subword fragmentation

# Define common Japanese particles to exclude from frequency analysis
common_particles = {"は", "の", "に", "が", "を", "で", "か", "と", "や", "も"}

# Load your Japanese question data (assuming you have a DataFrame with a 'question' column)
questions_ja = japanese_df['question'].tolist()

# Use MeCab to tokenize and filter out common particles
tokenized_texts = []
for question in questions_ja:
    # MeCab segmentation
    tokens = mecab.parse(question).strip().split()
    # Filter out common particles
    tokenized_texts.extend([token for token in tokens if token not in common_particles])

# Count the frequency of the filtered words
word_freq = Counter(tokenized_texts)

# Get the most common words (e.g., top 10)
most_common_words = word_freq.most_common(10)

# Print the most common words and their frequencies
print("Most common words:", most_common_words)

# Initialize the translator (using Googletrans for demonstration purposes)
translator = Translator()

# Translate the most common words and print results
translations = [(word, translator.translate(word, src='ja', dest='en').text) for word, _ in most_common_words]

# Print the translations
for word, translation in translations:
    print(f"{word} -> {translation}")
```

Other approach using the model translation built in tokenizers

```python
import pandas as pd
import string
from transformers import AutoTokenizer
from collections import Counter

# Load the parquet file containing the train split
splits = {'train': 'train.parquet'}
df = pd.read_parquet("hf://datasets/coastalcph/tydi_xor_rc/" + splits["train"])

# Filter questions based on language
languages = ['fi', 'ja', 'ru']  # Finnish, Japanese, Russian
questions_by_language = {lang: [] for lang in languages}

# Assuming 'question_text' is the column with questions and 'language' identifies the language
for lang in languages:
    # Filter the dataframe for each language
    lang_questions = df[df['lang'] == lang]['question']

    # Combine all questions into a single string per language
    questions_by_language[lang] = " ".join(lang_questions)

# Load the tokenizer for Finnish, Japanese, and Russian
tokenizers = {
    'fi': AutoTokenizer.from_pretrained('Helsinki-NLP/opus-mt-fi-en'),
    'ja': AutoTokenizer.from_pretrained('Helsinki-NLP/opus-mt-ja-en'),
    'ru': AutoTokenizer.from_pretrained('Helsinki-NLP/opus-mt-ru-en')
}

punctuation_set = set(string.punctuation)

for lang in languages:
    tokens = tokenizers[lang].tokenize(questions_by_language[lang])
    filtered_tokens = [t for t in tokens if not set(t).issubset(punctuation_set) or t in ['', '▁', '́']]

    # Count word frequencies
    word_counts = Counter(filtered_tokens)
    common_words = word_counts.most_common(10)

    print(f"Most common words in {lang}: {common_words}")
```
