```python
from datasets import load_dataset

dataset = load_dataset("coastalcph/tydi_xor_rc")

print(dataset)
```

```python
# Extract training and validation data for each language
# Training
finnish_train_questions = []
japanese_train_questions = []
russian_train_questions = []
english_train_contexts = []

for entry in dataset['train']:
    if entry['lang'] == 'fi':
        finnish_train_questions.append(entry['question'])
    elif entry['lang'] == 'ja':
        japanese_train_questions.append(entry['question'])
    elif entry['lang'] == 'ru':
        russian_train_questions.append(entry['question'])
    english_train_contexts.append(entry['context'])

# Validation
finnish_val_questions = []
japanese_val_questions = []
russian_val_questions = []
english_val_contexts = []

for entry in dataset['validation']:
    if entry['lang'] == 'fi':
        finnish_val_questions.append(entry['question'])
    elif entry['lang'] == 'ja':
        japanese_val_questions.append(entry['question'])
    elif entry['lang'] == 'ru':
        russian_val_questions.append(entry['question'])
    english_val_contexts.append(entry['context'])

# alternative method: finnish_train_data = dataset['train'].filter(lambda x: x['lang'] == 'fi')

# View the first example for each
print(finnish_train_questions[0])
print(japanese_train_questions[0])
print(russian_train_questions[0])
print(english_train_contexts[0])
print(finnish_val_questions[0])
print(japanese_val_questions[0])
print(russian_val_questions[0])
print(english_val_contexts[0])
```

```python
from datasets import Dataset

# Create datasets from raw train data
finnish_train_data = Dataset.from_dict({'text': finnish_train_questions})
japanese_train_data = Dataset.from_dict({'text': japanese_train_questions})
russian_train_data = Dataset.from_dict({'text': russian_train_questions})
english_train_data = Dataset.from_dict({'text': english_train_contexts})

print(finnish_train_data, "\n", finnish_train_data[0], "\n")
print(japanese_train_data, "\n", japanese_train_data[0], "\n")
print(russian_train_data, "\n", russian_train_data[0], "\n")
print(english_train_data, "\n", english_train_data[0], "\n")

# Create datasets from raw validation data
finnish_val_data = Dataset.from_dict({'text': finnish_val_questions})
japanese_val_data = Dataset.from_dict({'text': japanese_val_questions})
russian_val_data = Dataset.from_dict({'text': russian_val_questions})
english_val_data = Dataset.from_dict({'text': english_val_contexts})

print(finnish_val_data, "\n", finnish_val_data[0], "\n")
print(japanese_val_data, "\n", japanese_val_data[0], "\n")
print(russian_val_data, "\n", russian_val_data[0], "\n")
print(english_val_data, "\n", english_val_data[0], "\n")
```

```python
import ipadic
import fugashi
from transformers import AutoTokenizer

# Load tokenizers for each language
tokenizer_fi = AutoTokenizer.from_pretrained("TurkuNLP/gpt3-finnish-small")
tokenizer_ja = AutoTokenizer.from_pretrained("rinna/japanese-gpt2-medium")
tokenizer_ru = AutoTokenizer.from_pretrained("sberbank-ai/rugpt3medium_based_on_gpt2")
tokenizer_en = AutoTokenizer.from_pretrained("gpt2")

 # Use the end-of-sequence token as the padding token
tokenizer_fi.pad_token = tokenizer_fi.eos_token
tokenizer_ja.pad_token = tokenizer_ja.eos_token
tokenizer_ru.pad_token = tokenizer_ru.eos_token
tokenizer_en.pad_token = tokenizer_en.eos_token

# Tokenization function
def tokenize_data(dataset, tokenizer):
    return dataset.map(lambda x: tokenizer(x['text'], truncation=True, padding='max_length'), batched=True)

# Tokenize the datasets
finnish_tokenized_train_data = tokenize_data(finnish_train_data, tokenizer_fi)
japanese_tokenized_train_data = tokenize_data(japanese_train_data, tokenizer_ja)
russian_tokenized_train_data = tokenize_data(russian_train_data, tokenizer_ru)
english_tokenized_train_data = tokenize_data(english_train_data, tokenizer_en)

finnish_tokenized_val_data = tokenize_data(finnish_val_data, tokenizer_fi)
japanese_tokenized_val_data = tokenize_data(japanese_val_data, tokenizer_ja)
russian_tokenized_val_data = tokenize_data(russian_val_data, tokenizer_ru)
english_tokenized_val_data = tokenize_data(english_val_data, tokenizer_en)

# View the first tokenized example for each
print(finnish_tokenized_train_data[0])
print(japanese_tokenized_train_data[0])
print(russian_tokenized_train_data[0])
print(english_tokenized_train_data[0])
print(finnish_tokenized_val_data[0])
print(japanese_tokenized_val_data[0])
print(russian_tokenized_val_data[0])
print(english_tokenized_val_data[0])
```

**3. Perplexity calculation**

```python
from transformers import AutoModelForCausalLM

# Load models for each language
model_fi = AutoModelForCausalLM.from_pretrained("TurkuNLP/gpt3-finnish-small")
model_ja = AutoModelForCausalLM.from_pretrained("rinna/japanese-gpt2-medium")
model_ru = AutoModelForCausalLM.from_pretrained("sberbank-ai/rugpt3medium_based_on_gpt2")
model_en = AutoModelForCausalLM.from_pretrained("gpt2")
```

```python
import math
import torch
from tqdm import tqdm

# Function to compute perplexity on validation data
def compute_perplexity_on_dataset(model, tokenizer, dataset):
    total_loss = 0.0
    total_tokens = 0

    # Use tqdm to create a progress bar
    for entry in tqdm(dataset, desc="Computing perplexity"):
        inputs = tokenizer(entry['text'], return_tensors='pt', truncation=True, padding=True)
        with torch.no_grad():
            outputs = model(**inputs, labels=inputs["input_ids"])
        loss = outputs.loss.item()  # Cross-entropy loss
        num_tokens = inputs['input_ids'].size(1)  # Number of tokens in this input
        total_loss += loss * num_tokens  # Multiply loss by number of tokens
        total_tokens += num_tokens  # Keep track of total number of tokens

    # Average loss over all tokens
    average_loss = total_loss / total_tokens
    perplexity = math.exp(average_loss)  # Perplexity is exp of the average loss
    return perplexity
```

```python
print("Evaluating Finnish model:")
perplexity_fi_val = compute_perplexity_on_dataset(model_fi, tokenizer_fi, finnish_val_data)
print(f"\nPerplexity for Finnish: {perplexity_fi_val:.2f}")
```

```python
print("Evaluating Japanese model:")
perplexity_ja_val = compute_perplexity_on_dataset(model_ja, tokenizer_ja, japanese_val_data)
print(f"\nPerplexity for Japanese: {perplexity_ja_val:.2f}")
```

```python
print("Evaluating Russian model:")
perplexity_ru_val = compute_perplexity_on_dataset(model_ru, tokenizer_ru, russian_val_data)
print(f"\nPerplexity for Russian: {perplexity_ru_val:.2f}")
```

```python
print("Evaluating English model:")
perplexity_en_val = compute_perplexity_on_dataset(model_en, tokenizer_en, english_val_data)
print(f"\nPerplexity for English: {perplexity_en_val:.2f}")
```
